```python
tensor=tf.constant([98,40,32,22,12,39])
tensor
```

```python
#index of the min and max element of a tensor
min_element_index=tf.argmin(tensor)
max_element_index=tf.argmax(tensor)
min_element_index,max_element_index
```

```python
#value of min and max element of a tensor
min=tf.reduce_min(tensor)
max=tf.reduce_max(tensor)
min,max
```

```python
mat=tf.constant([[15,20,13],[84,45,36]])
mat
```

```python
#Find sum of a tensor
sum=tf.reduce_sum(mat)
sum
```

```python
#Find the sum along the axis

#Find the sum  row-wise
sum_0=tf.reduce_sum(mat,axis=0)
#Find the sum column-wise
sum_1=tf.reduce_sum(mat,axis=1)

#sum_0,sum_1
sum_0.numpy(),sum_1.numpy()
```

```python
mean = tf.reduce_mean(mat)
mean
```

```python
#Find the mean along the axis

#Find the mean row wise
mean_0=tf.reduce_mean(mat,axis=0)
#Find the mean column wise
mean_1=tf.reduce_mean(mat,axis=1)

#mean_0,mean_1
mean_0.numpy(),mean_1.numpy()
```

```python
#Find the variance of a tensor
import tensorflow_probability as tfp
variance=tfp.stats.variance(mat)
variance
```

```python
#Dot Product W*X
w=tf.constant([1,2,3])
x=tf.constant([4,5,6])
w,x
```

```python
dot_prod=tf.reduce_sum(tf.multiply(w,x))
dot_prod
```

```python
dot_product=tf.tensordot(w,x,axes=1)
dot_product
```

```python
#Square root of a tensor
z=tf.constant([16.0,25.0,36.0])
z
```

```python
Sqrt=tf.sqrt(z)
Sqrt
```

# Matrix Multiplication

```python
# Create a matrix Tensor
matrix1 =tf.constant([[1,2,3],[4,5,6],[11,12,45]])
matrix2 =tf.constant([[7,8],[9,10],[23,43]])
matrix3=tf.constant([[10,20,30],[33,42,87]])
matrix1,matrix2,matrix3
```

```python
matrix1.numpy()
```

```python
matmul1_2=tf.matmul(matrix1,matrix2)
matmul1_2
```

```python
#shortcut operator for multiplication
matrix1 @ matrix2
```

```python
#python matrix multiplication of matrices with different shapes
matrix1 * matrix3
```

```python
#python performs elementwise mulitiplication for that both matrix should be of same shape
matrix1 * matrix1
```

```python
#Matrix multiplication with different inner dimensions
matmul1_3=tf.matmul(matrix1,matrix3)
matmul1_3
```

```python
matrix1.shape,matrix3.shape
```

```python
matrix3_reshape=tf.reshape(matrix3,shape=(3,2))
matrix3_reshape
```

```python
tf.matmul(matrix1,matrix3_reshape)
```

```python
tf.reshape(matrix3,shape=(4,2))
```

#Concatenate Two Tensors: tf.concat()

```python
matrix1=[[3,4,6],[7,7,7]]
matrix2=[[1,2,3],[4,5,6]]
matrix1,matrix2
```

```python
#Concatenation of 2 matrix rowwise
tf.concat([matrix1,matrix2],axis=0)
```

```python
#concatenation of two matrices column-wise
tf.concat([matrix1,matrix2],axis=1)
```
